# Chapter 7

### Filter out missing data

```python
# Check the percentage of null values in each column sorted by descending order
df.isnull().sum().sort_values(ascending=False) / df.shape[0] * 100
```

```python
# Filter out the columns with atleast 90 percent null values
df1 = df.dropna(axis=1, thresh=df.shape[0]*0.90)
```

```python
# Check the percentage of null values in filtered dataframe
df1.isnull().sum().sort_values(ascending=False) / df.shape[0] * 100
```

```python
# Filter out null values
df1 = df1.dropna()
df1.isnull().sum().sort_values(ascending=False)
```

### Remove duplicates

```python
# Removing duplicates
df1.drop_duplicates()
```

### Convert between string and date time

```python
df1.info()
```

```python
df1['AccessionYear'] = df1['AccessionYear'].astype('str').replace(r'\.\d+$', '', regex=True)
df1['AccessionYear'] = df1['AccessionYear'].str[:4]
df1['AccessionYear']
```

```python
df1['AccessionYear'] = pd.to_datetime(df1['AccessionYear'])
```

```python
df1['AccessionYear']
```

```python
df1.info()
```

### Convert timestamps to periods and back

```python
df3 = pd.read_excel('CANDY-HIERARCHY-2015-SURVEY-Responses.xlsx')
```

```python
ts = df3.set_index('Timestamp')
ts
```

```python
# Converting the timestamp to period with minute frequency
ts_period = ts.to_period(freq='T')
ts_period
```

```python
# Converting back period data to timestamp with same minute frequency
ts_period.to_timestamp()
```
